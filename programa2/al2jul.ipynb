```julia
using Distributions
using Random

n = 100 #numero de cartas a coleccionar, se puede cambiar

i = 0 #  numero de intentos hasta obtener la coleccion
lst = collect(1:n) # lista de cupones
collection = Set{Int64}() # conjunto de la coleccion
num_cupones = 0  # inicializando cupon

x = DiscreteUniform(1,n) # inicializacion del tipo de distribucion

#algoritmo
while num_cupones < n
    i = i+1
    cupon = rand(x)
    #cupon = rand(lst[1:n])
    if in(cupon, collection) != true
        num_cupones += 1
        push!(collection, cupon) 
    end
end

print("intentos hasta obtener la coleccion")
print("\n")
print(i)
print("\n")
print("coleccion")
print("\n")
print(lst)
print("\n")
print("numero de cupones")
print("\n")
print(num_cupones)
print("\n")
print("como voy obteniendo la coleccion")
print("\n")
print(collection)
print("\n")
print("longitud de la coleccion obtenida")
print("\n")
print(length(collection))


# algoritmo en forma de funcion para experimentar con diferente numero de cartas e intentos
function collecionador(num)
    i = 0
    collection = Set{Int64}()
    numcup = 0
    x = DiscreteUniform(1,num)
    
    while numcup < num
        i = i+1
        cupon = rand(x)
        #cupon = lst[rand(1:end)]
        if in(cupon, collection) != true
            numcup += 1
            push!(collection, cupon) 
        end
    end 
    return i
end
```

Explicacion: 
 - la grafica azul es la funcion: $nIn(n)+O(n)$
 - la grafica amarilla : $n In (n)$
 - la grafica en rojo se obtiene ejecutando el algoritmo de coleccion de cupones para el numero x de cartas o cupones    solo una vez para x = 1,....100

```julia
using Plots
using Statistics



trial1  = zeros(0) # vector de intentos
yc1 = zeros(0) # vector para diferente numero de colecciones
x1 = collect(1:100) 

for j in x1 # coleccion i = 1, 2, .... n
    append!( yc1 , collecionador(j))
end

print(yc1)
```

```julia
# funcion n in (n) + O(n)
y11 = x1.*log.(x1)  + x1
# funcion n in (n) 
y21 = x1.*log.(x1)

# yc son los datos obtenidos de la simulacion de algoritmo colector
dat= [y11 y21 yc1]
plot(x1,dat)
```

### para 1000

```julia
using Plots
using Statistics


simulaciones = 30 # numero de simulaciones
trial  = zeros(0) # vector de intentos
yc = zeros(0) # vector para resultados promedio de obtener la coleccion, cuando la coleccion i, tal que i = 1,....n
x= collect(1:1000) 

for j in x # coleccion i = 1, 2, .... n
    for i= 1:simulaciones # numero de intentos o simulaciones
        append!( trial , collecionador(j) ) # vector donde se guardan intentos hasta obtener la coleccion i
    end
    append!( yc , mean(trial)) # se obtiene la media de los intentos o simulaciones
    trial  = zeros(0)
end

# funcion n in (n) + O(n)
y = x.*log.(x)  + x 
# funcion n in (n) 
y2 = x.*log.(x)
# yc son los datos obtenidos de la simulacion de algoritmo colector
dat= [y y2 yc]
plot(x,dat)
```
