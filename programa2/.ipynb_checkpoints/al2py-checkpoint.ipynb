```python
import numpy as np
n = 100 #numero de tazos totales
i = 0
collection= {}
num_cupones = 0
while num_cupones< n:
    i = i+1
    cupon = np.random.choice(list(range(1, n+1)))
    if (cupon in collection)!= True:
        num_cupones += 1
        collection[cupon]=True

print("intentos para alcanzar la coleccion")
print(i)
print("lista de la coleccion")
print(list(range(1, n+1)))
print("numero de cupones a coleccionar")
print(num_cupones)
print("como obtuve la coleccion")
print(collection)
```

```python
# funcion para comparar con los valores esperados teoricos

import random
import numpy as np
import matplotlib.pyplot as plt

# variables
cards = 100 # numero de cartas a coleccionar
Trials = []

# simulaciones
simulations = 4 

# recoleccion de tazos, numero de intentos 
def collect(cards):
    loops = 0
    Deck = []

    while len(Deck) < cards:
        card = random.randint(1, cards)
        #card = np.random.choice(list(range(1, n+1)))
        if card not in Deck:
            Deck.append(card)
        loops += 1

    return loops

# simulaciones
for x in range(simulations):
    Trials.append(collect(cards))

print("vector con intentos hasta obtener coleccion")
print(Trials)

# Algunos resultados de los intentos
print('Promedio:', np.average(Trials))
print('Mediana', np.median(Trials))
print('Minimo:', min(Trials))
print('Maximo:', max(Trials))
```

```python
n = 100 # numero de cartas a coleccionar, no puse mas por que se tarda mucho debido a la complejidad del algoritmo
        # entre mas incrementa por ejempo 10000 cupones el numero de intentos para obtener la coleccion crece y se debe
        # calcular ademas para 10, 0 30 intentosy ademas para 1, 2,3,....,10000

# coleccion de cartas
x = list(range(1, n+1))
xa = np.array(x)
#funcion n in(n)
y2 = xa*np.log(xa)
#funcion n in(n) + n
y = xa*np.log(xa) + xa

# obtencion de la coleccion en una corrida del algoritmo colector
y1 = np.zeros(n)
for i in xa:
    y1[i-1]= collect(i)
```

Explicacion: 
 - la grafica azul es la funcion: $nIn(n)+O(n)$
 - la grafica amarilla : $n In (n)$
 - la grafica en rojo se obtiene ejecutando el algoritmo de coleccion de cupones para el numero x de cartas o cupones    solo una vez para x = 1,....100

```python
# Graficando resultados
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x,y1,color= 'r')
plt.plot(x,y2,color= 'y')
plt.xlabel("numero de cupones a coleccionar")
plt.ylabel("intentos para alcanzar dicha coleccion")
plt.title("Grafica 1")
```
