## Verificacion de igualdad entre polinomios con diferentes formas

### Instrucciones
- copiar la informacion en text(3).txt
- Ejemplo de entrada:

48.84 0.86 -15.64 -0.5 1-0 

-2.0 3.0 2.2 -3.7

Que representa: 

$48.84 + 0.86x -15.64x^2 -0.5x^3 + x^4$

$(x-2)(x+3)(x+2.2)(x-3.7)$

otro ejemplo

1 2 3 4 5 6 7 8 -7 -6

```julia
using DelimitedFiles
using Distributions

f = open("text(3).txt")
list1 = readline(f)
list2 = readline(f)

x = split(list1)
y = split(list2)
xn = map(x->parse(Float64,x),x)
yn  = map(x->parse(Float64,x),y)

lst = collect(1:length(xn)*100.0)
close(f)

```
