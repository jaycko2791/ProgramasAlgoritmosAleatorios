```julia
using Distributions
using Random

n = 100 #numero de cartas a coleccionar, se puede cambiar

i = 0 #  numero de intentos hasta obtener la coleccion
lst = collect(1:n) # lista de cupones
collection = Set{Int64}() # conjunto de la coleccion
num_cupones = 0  # inicializando numero de cupones

x = DiscreteUniform(1,n) # inicializacion del tipo de distribucion

#algoritmo
while num_cupones < n
    i = i+1
    cupon = rand(x)
    #cupon = rand(lst[1:n])
    if in(cupon, collection) != true
        num_cupones += 1
        push!(collection, cupon) 
    end
end

print("intentos hasta obtener la coleccion")
print("\n")
print(i)
print("\n")
print("coleccion")
print("\n")
print(lst)
print("\n")
print("numero de cupones")
print("\n")
print(num_cupones)
print("\n")
print("como voy obteniendo la coleccion")
print("\n")
print(collection)
print("\n")
print("longitud de la coleccion obtenida")
print("\n")
print(length(collection))


# algoritmo en forma de funcion para experimentar con diferente numero de cartas e intentos
function collecionador(num)
    i = 0
    collection = Set{Int64}()
    numcup = 0
    x = DiscreteUniform(1,num)
    
    while numcup < num
        i = i+1
        cupon = rand(x)
        #cupon = lst[rand(1:end)]
        if in(cupon, collection) != true
            numcup += 1
            push!(collection, cupon) 
        end
    end 
    return i
end
```

```julia
# diferentes valores de cartas que se van a coleccionar

n1 = 10^1 
n2 = 10^2
n3 = 10^3
n4 = 10^4

C1 = range(-4,stop= 4,length= 20)
print(C1[3])
```

### Explicación

- para cada n, con n= 10, 100,1000 y 10000 que es la coleccion de cartas, se obtienen diferentes simulaciones del colector para observar la variable X, que es hasta cuantos intentos se hicieron para obtener la coleccion completa

- En un arreglo se guardan estos resultados o simulaciones y se cuentan aquellos que cumplen con $X <= nlog(n)+cn$ para diferentes valores de C, donde c esta en el rango de -4 a 4 

- para obtener la probabilidad, se cuentan los valores que cumplieron con $X <= nlog(n)+cn$ y se dividen entre el total de las simulaciones 

nota :  para obtener los resultados se puede hacer con un solo ciclo , pero lo hice así para que se puedan observar los resultados de  n1, n2, n3, n4

```julia
intentos = 10000 # simulaciones para coleccionar n1 cartas
atrials  = zeros(0) # vector de Ii (Intentos hasta obtener n1)
print(n1)

for i= 1:intentos # numero de intentos o simulaciones
        append!( atrials , collecionador(n1) ) # vector donde se guardan intentos hasta obtener la coleccion n1
end

P1 = zeros(0) # vector que cuenta la probabildad P(x <= nlog(n)+cn)  x es numero de intentos hasta obtener la coleccion
#y se realiza para diferentes valores de C 

# lo anterior se realiza contando los x menores a  nlog(n)+cn y dividiendolo entre el total de las simulaciones o intentos
for j = 1:length(C1)
    c = n1*log(n1) + C1[j]*n1
    p = 0
    for i in atrials
        if i <= c
            p = p+1
        end
    end     
    append!(P1, p/length(atrials))   
end

print(P1)
```
