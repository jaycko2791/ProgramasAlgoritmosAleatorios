```python
import numpy as np
import math


n = 101
c = 2 #variacion de numeros diferentes en el arreglo
a = np.random.randint(0,c*n, size=n)
print('arreglo desordenado donde se quiere encontrar la media')
print(a)

def findmedian1(a):
    sizeR = math.ceil(n**(3/4))
    #print(sizeR)
    R = []
    i = 0
    while i <= sizeR:
        i = i+1
        R.append(np.random.choice(a))
    
    Ra = np.array(R)
    Ra.sort()
    d =  Ra[math.floor(1/2*(n**(3/4)) - math.sqrt(n))+1]
    u =  Ra[math.ceil(1/2*(n**(3/4)) + math.sqrt(n))+1]
    Ld = []
    Lu = []
    C = []
    
    for i in a:
        if i > u:
            Lu.append(i)
        elif i < d:
            Ld.append(i)
        elif i>= d and i<= u:
            C.append(i)
    
    if len(Ld)>n/2 or len(Lu)>n/2:
        print("fallaste")
    elif len(C) <= 4*(n**(3/4)):
        Ca = np.array(C)
        Ca.sort()
        mediana=Ca[math.floor(n/2) - len(Ld) +1]
        print('mediana')
        print(mediana)   
    else: 
        print("fallaste")
        
def findmedian2(a,k):
    if len(a)<=1:
        return(a[0])
    
    print('d')
    print(a)
    
    menp = []
    mayp = []
    p = np.random.choice(a)
    for i in a:
        if i > p:
            mayp.append(i)
            myp = np.array(mayp)
            
            
        elif i < d:
            menp.append(i)
            mnp = np.array(menp)
    
    if len(menp)>k:
        findmedian2(np.array(menp),k)
    
    elif len(mayp)>k:
        findmedian2(np.array(mayp),k-len(menp)-1)
    
    else: 
        return p



findmedian1(a)

print('encontrar la media con r-select')
median = findmedian2(a,math.floor(n/2))
print(median)

        
        
print('resultados con algoritmo o(nlogn)')    
a.sort()
print('arreglo ordenado')
print(a)
print('mediana')
print(a[51])
```
